```python
import spym
import numpy as np
import os

# 加载 SM4 文件
def inspect_sm4_file(filename):
    # 读取 .SM4 文件
    file = spym.load(filename)
    
    # 提取所有通道的信息
    channels = list(file.data_vars)
    print(f"Found {len(channels)} channels:\n")
    
    for channel in channels:
        data = file[channel].data
        print(f"Channel: {channel}")
        print(f"Shape: {data.shape}")
        print("-")

    # 用户选择通道并导出为 txt 文件
    selected_channels = input("Enter the channel names to export (comma-separated): ").split(',')
    selected_channels = [ch.strip() for ch in selected_channels]

    output_dir = os.path.dirname(filename)
    for selected_channel in selected_channels:
        if selected_channel in channels:
            data = file[selected_channel].data
            output_path = os.path.join(output_dir, f"{selected_channel}.txt")
            np.savetxt(output_path, data, delimiter='\t')
            print(f"Channel '{selected_channel}' data exported to {output_path}")
        else:
            print(f"Channel '{selected_channel}' not found.")

# 示例用法：
inspect_sm4_file(r"D:\系统默认\文档\实验数据\STM\TbMnSn\mapping\TbMnSn_0068.sm4")
```
